```python
#Compute x and y coordinates for point
x=np.arange(0, 3 * np.pi, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#Set up a subplot grid that has height
#and set the first such subplot as act
plt.subplot(2, 1, 1)
#Make the first plot as active
plt.plot(x,y_sin)
plt.title('Sine')
#Make the second plot as active
plt.subplot(2, 1, 2)
plt.plot(x,y_cos)
plt.title('Cosine')
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [5,8,10]
y = [12,16,6]

x2 = [6,7,9,11]
y2 = [6,10,15,7]
plt.bar(x, y, align = 'center')
plt.bar(x2, y2, color = 'm', align = 'center')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```
